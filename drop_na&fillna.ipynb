# dropna & fillna

```python
import pandas as pd
var = pd.read_csv("annual.csv")
var
```

```python
var.dropna()
```

```python
var.dropna(axis=0)
```

```python
var.dropna(how="any")
```

```python
var.dropna(how="all")
```

```python
var.dropna(subset=["Industry_code_NZSIOC"])
```

```python
var.dropna(inplace=True)
```

```python
var.dropna(thresh=1)
```

```python
var.fillna("python")
```

```python
var.fillna({"Industry_aggregation_NZSIOC":"python","Year":2022})
```

```python
var.fillna(method="ffill")
```

```python
var.fillna(method="bfill", axis=1)
```

```python
var.fillna(12,inplace=True)
```

```python
var.fillna("python",limit=2)
```
